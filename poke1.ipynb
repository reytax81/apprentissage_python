```python
# beaucoup de data sont dispo sur kaggle https://www.kaggle.com/rounakbanik/pokemon
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/anisayari/Youtube-apprendre-python-intelligence-artificielle/master/%238%20-%20Data%20Visualisation%20(Seaborn%2C%20Plotly)/pokemon.csv')
df.head(6)
```

```python
#describe
df.describe()
```

```python
#print type numeric
df.select_dtypes(include=['int64','float64']).columns
```

```python
#data preparation

df.columns = df.columns.str.upper() # titre des colonnes en majuscule
df['TYPE1'] = df['TYPE1'].str.capitalize() # premiere lettre en maj
df['TYPE2'] = df['TYPE2'].str.capitalize() # idem pour type 2
df['TYPE2'].fillna(value='none', inplace=True) #permet d'eviter certaine erreur du au none et null 
df.head()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set (style="whitegrid")
```

```python
#courbe
# graphic point de defense par rapport au point d'attaque
sns.lineplot(x="DEFENSE", y="ATTACK", data=df.sort_values(by='DEFENSE'))
```

```python
# montre la correlation des info
df.corr()
```

```python
%time
plt.figure(figsize=(20,10))

sns.heatmap(df.corr(),cmap=sns.color_palette("RdBu_r", 7))
```

```python
#heatmap / correlation

plt.figure(figsize=(10,10))

sns.heatmap(
    df[df['TYPE2']!='none'].groupby(['TYPE1','TYPE2']).size().unstack(),
    linewidths=1,
    annot=True, cmap=sns.color_palette("RdBu_r", 7)
)
```

```python
#nuage de point relplot
plt.figure(figsize=(20,20)) # modifie la taille du graph

sns.relplot(x='DEFENSE', y='ATTACK', hue='GENERATION', size='WEIGHT_KG',
            sizes=(40,400), alpha=.5, palette="muted",
            height=6, data=df[df['GENERATION'].isin([1,7])])
```

```python
# boite a moustache - boxplot
# plot the orbital period with horizontal box

plt.figure(figsize=(20,20))
sns.boxplot(x='GENERATION', y='WEIGHT_KG',hue='IS_LEGENDARY', data=df, palette="vlag")
```

```python
df[(df['GENERATION']==7) & (df['WEIGHT_KG']>800)]
```

```python
#histogramme
sns.countplot(x='GENERATION', data=df)
```

```python
#DIAGRAMME EN VIOLON
sns.set (style="whitegrid")

g = sns.catplot(x='GENERATION', y='HP', hue='IS_LEGENDARY', data=df,
                height=6, kind='bar', palette='muted')
g.despine(left=True)
g.set_ylabels('HP')
```

```python
fig = plt.figure(figsize=(7,7))

colours = ["cyan","green"]
pokeLeg = df[df['IS_LEGENDARY']==True]
pokeNon = df[df['IS_LEGENDARY']==False]

legDist = [pokeLeg['NAME'].count(),pokeNon['NAME'].count()]
legPie = plt.pie(legDist,
                 labels= ['legendary', 'Non legendary'],
                 autopct = '%1.1f%%',
                 shadow = True,
                 colors = colours,
                 startangle = 45,
                 explode=(0,0.1))
```

```python
dict (
    r=[1,5,2,2,3],
    theta=['processing cost', 'mechanical properties', 'chemical stability',
           'thermal stability', 'device integration'])
```

```python
import plotly.express as px
fig = px.line_polar(pd.DataFrame(dict (r=list (dict_tmp.values()),theta=list (dict_tmp.keys()))), r= 'r', theta= 'theta', line_close=True)
fig.update_layout(
    height=800,
    title_text=df.iloc[5].NAME
)
fig.show()
```
